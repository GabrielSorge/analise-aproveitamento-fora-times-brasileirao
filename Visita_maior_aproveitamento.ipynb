```python
import pandas as pd

df = pd.read_csv("matches-2003-2025.txt", sep = ";", encoding = "latin1")

df.shape
```

```python
df["#DATE"] = pd.to_datetime(df["#DATE"])
df["#DATE"]
```

```python

def verifica_resultado(score):
  placar = score.split("x")

  casa = int(placar[0])
  visitante = int(placar[1])

  if casa > visitante:
    resultado_visitante = "derrota"
  elif casa == visitante:
    resultado_visitante = "empate"
  else:
    resultado_visitante = "vitoria"

  return resultado_visitante
```

```python
df["RESULTADO_VISITANTE"] = df["SCORE"].apply(verifica_resultado)
df[["RESULTADO_VISITANTE", "AWAY-TEAM"]]
```

```python
jogos_fora = df.groupby("AWAY-TEAM")["AWAY-TEAM"].count()
print(jogos_fora)
```

```python
def pontos_partida(resultado):
  if resultado == "derrota":
    return 0
  elif resultado == "empate":
    return 1
  elif resultado == "vitoria":
    return 3
```

```python
df["PONTOS"] = df["RESULTADO_VISITANTE"].apply(pontos_partida)
```

```python
pontos = df.groupby("AWAY-TEAM")["PONTOS"].sum()
print(pontos)
```

```python
pontos_possiveis = jogos_fora * 3
print(pontos_possiveis)
```

```python
aproveitamento = (pontos / pontos_possiveis) * 100
print(round(aproveitamento,2).sort_values(ascending=False))
```

```python
jogos_fora_sem_outliers = jogos_fora[jogos_fora > 100]
pontos_possiveis_sem_outliers = jogos_fora_sem_outliers * 3
aproveitamento_sem_outliers = (pontos / pontos_possiveis_sem_outliers) * 100

aproveitamento_sem_outliers.dropna().sort_values(ascending=False).round(2)
```
